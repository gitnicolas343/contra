```python
# test_secop2.py

from listas_negras_client import consultar_secop2_por_nombre

def main():
    nombre = "TRANSPORTE EL LIBERTADOR SAS"  # el ejemplo que pasaste

    print(f"Consultando SECOP II para: {nombre!r}")
    df = consultar_secop2_por_nombre(nombre, limite=50000)

    print("\nShape del DataFrame:", df.shape)
    print("\nColumnas:", list(df.columns))

    # Si encontró algo, mostramos las columnas clave
    if not df.empty:
        cols_mostrar = [
            "nombre_proveedor_objeto_de",
            "codigo_entidad_creadora",
            "nombre_entidad_creadora",
            "as_codigo_proveedor_objeto",
            "valor",
            "valor_pagado",
            "fecha_evento",
            "numero_de_acto",
            "tipo_de_sancion",
            "estado",
        ]
        cols_mostrar = [c for c in cols_mostrar if c in df.columns]

        print("\nPrimeras filas filtradas:")
        print(df[cols_mostrar].head(10))
    else:
        print("\nNo se encontraron registros para ese nombre (tras filtrar en pandas).")

if __name__ == "__main__":
    main()
```
